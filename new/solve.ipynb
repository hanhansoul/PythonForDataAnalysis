```python
import pandas as pd

input_excel = pd.read_excel('input.xlsx', sheet_name=1, dtype=str)
output_excel = pd.read_excel('output.xlsx', dtype=str)
```

```python
input_excel.tail()
```

```python
output_excel.tail()
```

```python
in_df = pd.DataFrame({'批次': input_excel['批次'], '序号': input_excel['序号']})
in_df = in_df.drop_duplicates()
in_df.tail()
```

```python
out_df = pd.DataFrame({'批次': output_excel['批次'], '序号': output_excel['序号']})
```

```python
out_df
```

```python
res = pd.merge(out_df, in_df, left_on='批次', right_on='批次', how='left')
```

```python
res
```

```python
res.drop_duplicates(subset='批次', inplace=True)
```

```python
len(res)
```

```python
writer = pd.ExcelWriter('output2.xlsx')
res.to_excel(writer, index=False)
writer.save()
```
