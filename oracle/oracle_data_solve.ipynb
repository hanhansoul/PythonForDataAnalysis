```python
# -*- coding:utf-8 -*-

import pandas as pd
import csv
```

```python
columns_mappings = {'GI_DOCUMENT_NO': '发料单号', 
                   'VPN': '件号', 
                   'SN': '序号', 
                   'QTY': '数量',
                   'CHARG': '批次',
                   'INPUT_DATE_A': '领料日期',
                   'ACNO': '飞机号',
                   'MAINTENANCE_BASE': '维修基地',
                   'BILL_NO': '拆换单号',
                   'REPLACE_DATE': '拆换日期',
                   'REPLACE_TYPE': '工作类型',
                   'TL_QTY': '退料数量',
                   'REMOVE_VPN': '拆下件号',
                   'REMOVE_SN': '拆下序号',
                   'IF_HCJ1': '回仓标识',
                   'DEAL_STATUS': '库房签收'}

data = pd.read_csv('input1.csv', dtype=str, encoding='gbk')
data.sort_values(by=['REPLACE_TYPE', 'REPLACE_DATE'], inplace=True)
data['DEAL_STATUS'].replace(dict(A='未处理', B='已签发'), inplace=True)
data['REPLACE_TYPE'].replace({'01': '拆装', '02': '单装', '03': '单拆', '05': '领出串件退料', 
                              '06': '领出未用退料', '07':'领出升级退料', '08': '改装包器材退料', 
                              '99': '检测', '09': '探伤后退料'}, inplace=True)
data['QTY'] = data['QTY'].apply(lambda x: x.split('.')[0])
data['MAINTENANCE_BASE'].replace({'CD01': '成都基地'}, inplace=True)
data.fillna('', inplace=True)
data.rename(columns=columns_mappings, inplace=True)
data
```
