```python
import numpy as np
import pandas as pd

filename = 'input.xlsx'
df = pd.read_excel(filename, sheet_name=1)
# df[['MEN', 'TASK', 'ACCESS', 'PREP']] = df[['MEN', 'TASK', 'ACCESS', 'PREP']].fillna(0)
df[['MEN', 'TASK', 'ACCESS', 'PREP']] = df[['MEN', 'TASK', 'ACCESS', 'PREP']].apply(pd.to_numeric, errors='coerce')
df[['MEN', 'TASK', 'ACCESS', 'PREP']] = df[['MEN', 'TASK', 'ACCESS', 'PREP']].fillna(0)
df['SCORE'] = df.apply(lambda x: x['MEN'] * x['TASK'] + x['ACCESS'] + x['PREP'], axis=1)

df.sort_values(by='TASKNUMBER')
sdf = df[['TASKNUMBER', 'SCORE']].groupby(by='TASKNUMBER').sum().reset_index()
sdf.to_excel('output' + filename)
```
