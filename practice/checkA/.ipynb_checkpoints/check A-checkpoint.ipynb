```python
import pandas as pd
from datetime import date
import numpy as np
```

```python
input_filename = 'input.xlsx'
# type_dict = dict(HVNUMAP=str, HVAPPA4=np.int32, HVAPPMM=np.int32, HVAPPJJ=np.int32, 
#                 HVAPPDOC1=str, HVAPPDOC2=str, HVAPPH=np.int32, HVAPPC=np.int32, 
#                 HVPLAA4=np.int32, HVPLAMM=np.int32, HVPLAJJ=np.int32, HVPLAH=np.int32, 
#                 HVPLAC=np.int32)
type_dict = dict(HVNUMAP=str, HVAPPL=str, HVAPPA4=np.int32, HVAPPMM=np.int32, HVAPPJJ=np.int32, 
                HVAPPDOC1=str, HVAPPDOC2=str, HVAPPV=str, HVAPPH=np.int32, HVAPPC=np.int32)

frav = pd.read_excel('FRAV.xlsx')
checks = pd.read_excel(input_filename, dtype=type_dict)
# checks = checks.replace('nan', '')
# checks = checks.applymap(lambda x: x.strip())
```

```python
checks.columns
```

```python
def date_join(years, months, days):
    ll = []
    for y, m, d in zip(years, months, days):
        if y != '' and m != '' and d != '':
            ll.append(str(int(y)) + '/' + str(int(m)) + '/' + str(int(d)))
        else:
            ll.append('')
    return pd.Series(ll)

def generate_date(years, months, days):
    ll = []
    for y, m, d in zip(years, months, days):
        if y != '' and m != '' and d != '':
            ll.append(date(int(y), int(m), int(d)))
        else:
            ll.append(None)
    return pd.Series(ll)
```

```python
frav = frav.applymap(lambda x: x.strip())
checks['AVOFFICIEL'] = checks.merge(frav[['AVNUMAP', 'AVOFFICIEL']], how='left', left_on='HVNUMAP', right_on='AVNUMAP')['AVOFFICIEL']
```

```python
# 按导出列名重构fmpg
d = {'AC': checks['AVOFFICIEL'].str.strip(),
     'ACTYPE': checks['HVAPPL'].str.strip(),
     'DOCUMENT': checks['HVAPPDOC1'].str.strip() + checks['HVAPPDOC2'].str.strip(),
     'DATE': date_join(checks['HVAPPA4'], checks['HVAPPMM'], checks['HVAPPJJ']),
     'TIME': checks['HVAPPH'],
     'CYCLE': checks['HVAPPC'],
     'CHECK': checks['HVAPPV'],
     'SORT_DATE': generate_date(checks['HVAPPA4'], checks['HVAPPMM'], checks['HVAPPJJ'])
    }

new_checks = pd.DataFrame(d, columns=['AC', 'ACTYPE', 'DOCUMENT', 'DATE', 'TIME', 'CYCLE', 'CHECK', 'SORT_DATE'])
```

```python
new_checks = new_checks.sort_values(by=['AC', 'SORT_DATE'], ascending=[True, False])
```

```python
new_checks[new_checks['AC'] == 'B-6518']
```

```python
c = ['AC', 'ACTYPE', 'DOCUMENT', 'DATE', 'TIME', 'CYCLE', 'CHECK',
       'SORT_DATE', 'TIME_DELTA', 'CYCLE_DELTA']

blank = pd.DataFrame(data=[[None] * 10], columns=c)
blank
```

```python
res = None
for k, df in new_checks.groupby(['AC']):
    df['TIME_DELTA'] = df['TIME'] - df['TIME'].shift(-1)
    df['CYCLE_DELTA'] = df['CYCLE'] - df['CYCLE'].shift(-1)
    if res is None:
        res = pd.DataFrame(data=df)
    else:
        res = res.append(blank, ignore_index=True)
        res = res.append(df, ignore_index=True)
res = res.replace('nan', '')
res = res.fillna('')
res.to_excel('output.xlsx', columns=['AC', 'ACTYPE', 'DOCUMENT', 'SORT_DATE', 'TIME', 'CYCLE', 'CHECK', 
                                    'TIME_DELTA', 'CYCLE_DELTA'])
res
```
