```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_excel('6679.xls')
```

```python
df = df.dropna()
```

```python
df['Performance date'] = df['Performance date'].apply(lambda x: str(x.year).zfill(4) + str(x.month).zfill(2) + str(x.day).zfill(2))
df['AC'] = df['AC'].apply(lambda y: 'B-' + str(y))
```

```python
df.to_excel('6419output.xlsx')
```
