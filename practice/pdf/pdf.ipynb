```python
import pdfplumber
import pandas as pd
```

```python
pdf = pdfplumber.open('TM LIST.pdf')
table_settings = {
    "vertical_strategy": "explicit",
    "explicit_vertical_lines": [50, 150, 230, 550, 680, 790],
    "horizontal_strategy": "lines",
}

chart = None
for page in pdf.pages:
    table = page.extract_table(table_settings)
    l = pd.DataFrame(table[1:], columns=table[0])
    if chart is None:
        chart = pd.DataFrame(l)
    else:
        chart = chart.append(l)

chart
```

```python
chart['PROPER SHIPPING NAME']
```

```python
import re
pattern = re.compile('\s+')
chart['PROPER SHIPPING NAME'] = chart['PROPER SHIPPING NAME'].apply(lambda x: pattern.sub(' ', x.strip()))
```

```python
writer = pd.ExcelWriter('output.xlsx')
chart.to_excel(writer)
writer.save()
```
