```python
import pandas as pd

input1 = pd.read_excel('input1.xlsx', dtype=str)
input2 = pd.read_excel('input2.xlsx', sheet_name=list(range(4)))
```

```python
input_list = []
for k in input2:
    input_list.append(input2[k])
input0 = pd.concat(input_list)
input0.reset_index(drop=True, inplace=True)

input0 = input0.drop('允装件号', axis=1).join(input0['允装件号']
          .str.split(',', expand=True)
          .stack()
          .reset_index(level=1, drop=True)
          .rename('允装件号')
          .str.strip())
input0.reset_index(drop=True, inplace=True)
input0.drop_duplicates(inplace=True)

input1.drop('ID', axis=1, inplace=True)
```

```python
res = pd.merge(input1, input0[['构型机型', '允装件号', 'FIN CODE']], 
               left_on=['机型', '件号'], right_on=['构型机型', '允装件号'], how='left')
res['是否重复'] = res.duplicated(subset=['机型', 'MSN', '件号', '描述', '序号'], keep=False)
res.drop_duplicates(subset=['机型', 'MSN', '件号', '描述', '序号'], inplace=True)
res
```

```python
writer = pd.ExcelWriter('output.xlsx')
res.to_excel(writer, index=False)
writer.save()
```
