```python
import pandas as pd

df = pd.read_excel("result.xlsx", header=None)
df
```

```python
df[3] = df[3].apply(lambda x: ',' + x)
data = df.groupby(by=[0, 1, 2]).sum()
data
```

```python
data[3] = data[3].apply(lambda x: x[1:])
```

```python
writer = pd.ExcelWriter('output.xlsx')
data.to_excel(writer)
writer.save()
```
