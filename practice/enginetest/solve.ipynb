```python
import pandas as pd

df = pd.read_excel("input.xls", dtype=str)
df
```

```python
def ff(arr):
    if len(arr) == 2:
        arr.insert(1, '')
    return arr
    
mid = pd.DataFrame(df['所在单位'].str.split('-', expand=True))
mid
```

```python
res = pd.concat([df, mid], axis='columns')
writer = pd.ExcelWriter('output.xlsx')
res.to_excel(writer)
writer.save()
```

```python
person = res[['序号', '姓名', '授权号（工号）', '授权人类型', '执照编号', '专业', 0, 1, 2]]
person.drop_duplicates(inplace=True)
writer = pd.ExcelWriter('person.xlsx')
person.to_excel(writer)
writer.save()
```
