```python
import os
import pandas as pd
from collections import defaultdict

path = '项目蓝图'
d = defaultdict(list)
for item in os.listdir(path):
    s = item.split('-')
    d['MAIN'].append(s[0])
    d['TITLE'].append('-'.join((str(x) for x in s[1: -1])))
    d['LIST'].append((s[-1] if '.' not in s[-1] else s[-1].split('.')[0]).replace('_', '、'))

df = pd.DataFrame(columns=['MAIN', 'TITLE', 'LIST'])

d
```

```python
path = '项目蓝图'
items = []
for item in os.listdir(path):
    s = item.split('-')
    items.append(dict(
        MAIN=s[0], 
        TITLE='-'.join((str(x) for x in s[1: -1])), 
        LIST=(s[-1] if '.' not in s[-1] else s[-1].split('.')[0]).replace('_', '、')
    ))

df = pd.DataFrame(items, columns=['MAIN', 'TITLE', 'LIST'])
df.to_excel('list.xlsx')
```
