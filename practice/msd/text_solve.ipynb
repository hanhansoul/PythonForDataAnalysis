```python
import pandas as pd

reader = pd.read_csv('input.txt', sep='\t', encoding='utf-8', chunksize=1000)
```

```python
sum = 0
for chunk in reader:
    sum += chunk['HAZ'].count()
sum
```

```python
df = None
for chunk in reader:
    if df is None:
        df = pd.DataFrame(chunk[pd.notna(chunk['HAZ'])])
    else:
        df = pd.concat([df, chunk[pd.notna(chunk['HAZ'])]])
#     break
    
# df['HAZ']
df.reset_index(drop=True, inplace=True)
df
```
