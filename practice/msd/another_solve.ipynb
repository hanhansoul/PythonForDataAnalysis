```python
import pandas as pd

reader = pd.read_csv('input.txt', sep='\t', encoding='utf-8', chunksize=1000)

df = None
for chunk in reader:
    if df is None:
        df = pd.DataFrame(chunk[pd.notna(chunk['HAZ'])])
    else:
        df = pd.concat([df, chunk[pd.notna(chunk['HAZ'])]])
        
df.reset_index(drop=True, inplace=True)
```

```python
export_df = pd.read_excel('export.xlsx')
export_res = pd.merge(df[['PNR', 'HAZ']], export_df[['物料', '危险品编码']], left_on='PNR', right_on='物料', how='right')
res = export_res[pd.notna(export_res['HAZ']) & pd.isna(export_res['危险品编码'])]
```
